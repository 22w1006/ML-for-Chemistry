  - è¦æ±‚: ç´¢å¼•å€¼å¤„äºåŒºé—´`[0, 983)`å†…æ—¶è¿”å›é˜³æ€§æ ·æœ¬çš„å¯¹åº”æ–‡ä»¶, å¤„äºåŒºé—´`[983, 1966)`æ—¶è¿”å›é˜´æ€§æ ·æœ¬çš„å¯¹åº”æ–‡ä»¶, åºå·ä»¥Pythonå†…ç½®åº“å‡½æ•°[`os.listdir()`](https://docs.python.org/3/library/os.html#os.listdir)æ’å‡ºçš„é¡ºåºç»™å®š.

### æç¤º
- ç±»å±æ€§çš„è®¿é—®éœ€è¦åŠ å‰ç¼€`self.`, ä¾‹å¦‚`self.device`ã€`self.transform`ç­‰.
- å›¾ç‰‡æ–‡ä»¶çš„è¯»å…¥å¯ç”¨[`PIL.Image.open()`](https://pillow.readthedocs.io/en/stable/reference/Image.html#PIL.Image.open)å‡½æ•°å®Œæˆ. è¯»å…¥å, ä½ æ‰€å¾—åˆ°çš„`PIL.Image.Image`å¯¹è±¡å¯ç”±[`torchvision.transforms.ToTensor()`](https://pytorch.org/vision/stable/generated/torchvision.transforms.ToTensor.html#torchvision.transforms.ToTensor)æ–¹æ³•è½¬æ¢æˆ`torch.tensor`æ ¼å¼.
- æœ¬æ¬¡ä½œä¸šå»ºè®®å¤§å®¶åœ¨å…è´¹çš„`c2_m4_cpu`æœºå™¨ä¸Šè°ƒè¯•å®Œä»£ç å, åˆ©ç”¨GPUèµ„æºå®Œæˆæ­£å¼çš„è¶…å‚æ•°ä¼˜åŒ–ä¸æ¨¡å‹è®­ç»ƒ(é‡æ–°è¿æ¥æœºå™¨å, ç‚¹å‡»restart kernel and run all cellsæŒ‰é’®). ä½œä¸šå†…æµ‹æ—¶ä½¿ç”¨çš„GPUè®¾å¤‡æ˜¯`c12_m46_1 * NVIDIA GPU B`, æ•ˆç‡çº¦æœ‰4å€çš„æå‡(~20s / epoch, é¢„è®¡ä»å¤´è¿è¡Œnotebookçš„æ€»èŠ±è´¹çº¦45minã€5.25å…ƒ).
  - æ•°æ®é›†çš„åˆå§‹åŒ–æ–¹æ³•å«æœ‰å‚æ•°`device`, æ‰€ä»¥, åœ¨`__getitem__()`æ–¹æ³•ä¸­, ä½ éœ€è¦ä¿è¯ä½ è¿”å›çš„`tensor`å¯¹è±¡è¢«æ¬è¿åˆ°äº†æ­£ç¡®çš„`self.device`å±æ€§ä¸Š. å¯ä»¥åˆ©ç”¨[`tensor.to()`](https://pytorch.org/docs/stable/generated/torch.Tensor.to.html#torch-tensor-to)æ–¹æ³•.
  - æ¨¡å‹åœ¨æ¯æ¬¡å®Œæˆåˆ›å»ºåä¹Ÿåº”å½“æ¬è¿åˆ°`device`ä¸Š, ç¨åè¯¦è¿°.
  - è¯¾ç¨‹é¢„ç®—æœ‰é™, è¯·ç¡®ä¿ä½ çš„ä»£ç ä¸éœ€è¦å¤§ä¿®åå†è¿æ¥GPUèµ„æº, å¹¶è°¨æ…ä½¿ç”¨.
- å¦‚æœæƒ³åœ¨`device`ä¸ŠåŸä½æ„é€ æ–°çš„`tensor`å¯¹è±¡, è¯·ä½¿ç”¨å°å†™çš„t, å³`torch.tensor(..., device=self.device)`è€Œé`torch.Tensor(..., device=self.device)`, åè€…[ä¸æ”¯æŒGPUä¸Šç›´æ¥åˆ›å»º](https://discuss.pytorch.org/t/cannot-construct-tensor-directly-on-gpu-in-torch-1-10-1/153751).

```python
import os
import sys
from torch.utils.data import Dataset
from PIL import Image

class PizzaOrNotDataset(Dataset):
    def __init__(self, root_dir: str, transform=None, device="cpu"):
        self.positive_idxs = os.listdir(os.path.join(root_dir, "pizza"))
        self.negative_idxs = os.listdir(os.path.join(root_dir, "not_pizza"))
        self.root_dir = root_dir
        self.transform = transform
        self.device = device
    def __len__(self):
        ### BEGIN YOUR SOLUTION ###
        return self.n_positive + self.n_negative
        ### END YOUR SOLUTION ###
    def __getitem__(self, idx: int):
        ### BEGIN YOUR SOLUTION ###
        positive_dir = 'pizza/'
        negative_dir = "not_pizza/"
        
        if 0 <= idx < self.n_positive:
            idx = idx
            img = self.positive_idxs[idx]
            X = self.transform(Image.open(self.root_dir+positive_dir+self.positive_idxs[idx]))
            return X, torch.tensor(1, dtype=torch.float)
        elif self.n_positive <= idx < self.n_positive+self.n_negative:
            idx = idx - self.n_positive
            img = self.negative_idxs[idx]
            X = self.transform(Image.open(self.root_dir+negative_dir+self.negative_idxs[idx]))
            return X, torch.tensor(0, dtype=torch.float)
        else:
            print("Invalid index, system exit\n")
            sys.exit(-1)
            
        
        
        ### END YOUR SOLUTION ###
    @property
    def n_positive(self):
        return len(self.positive_idxs)
    @property
    def n_negative(self):
        return len(self.negative_idxs)
```

$\quad$æˆ‘ä»¬æ­å»ºçš„æ¨¡å‹æ¶æ„å¦‚ä¸‹å›¾æ‰€ç¤º. è¿™æ˜¯ä¸€ä¸ªå·ç§¯ç½‘ç»œ+å…¨è¿æ¥ç½‘ç»œçš„æ¶æ„, å…¶ä¸­, å·ç§¯æ¨¡å—`ConvModule`åŒ…å«ä¸¤ç»„å·ç§¯-æœ€å¤§æ± åŒ–æ“ä½œ, å¹¶ä¸”ç¬¬äºŒæ¬¡çš„å·ç§¯æ ¸æœ‰æ‰€æ‰©å¼ , ä»¥æå–æ›´å¹¿åŸŸçš„åƒç´ ä¿¡æ¯. æ¨¡å‹çš„æœ€ç»ˆè¾“å‡ºç”±Sigmoidå‡½æ•°æ¿€æ´», å¹¶è§£é‡Šä¸º(äºŒåˆ†ç±»é—®é¢˜ä¸­çš„)é˜³æ€§æ¦‚ç‡. å›¾ä¸­é»˜è®¤é‡‡ç”¨3x3å·ç§¯æ ¸å’Œ1024å¤§å°çš„éšè—å±‚, è¿™ä¸¤ä¸ªå€¼å°†åœ¨ç¨åä½œä¸ºæ¨¡å‹è¶…å‚æ•°è¿›è¡Œè°ƒæ•´.

$\quad$ä»¥ä¸‹æ˜¯å·ç§¯æ¨¡å—`ConvModule`çš„ä»£ç å®ç°.

```python
from torch import nn
from torch.nn import functional as F

class ConvModule(nn.Module):
    def __init__(self, kernel_size=3):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 8, kernel_size=kernel_size, dilation=1, padding="same")
        self.pooling1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(8, 16, kernel_size=kernel_size, dilation=3, padding="same")
        self.pooling2 = nn.MaxPool2d(kernel_size=2, stride=2)
    def forward(self, X):
        X = X.cuda() if torch.cuda.is_available() else X
        X = self.pooling1(F.relu(self.conv1(X)))
        X = self.pooling2(F.relu(self.conv2(X)))
        return X
```

$\quad$**ä»»åŠ¡2**: æ ¹æ®å·²ç»™å‡ºçš„`ConvModule`æ¨¡å—ä»£ç ä»¥åŠå›¾ç¤ºæ¨¡å‹æ¶æ„, å®Œæˆæ¨¡å‹å­ç±»`PizzaOrNotModel`çš„ç¼–å†™. ä½ éœ€è¦å®ç°å¦‚ä¸‹ä¸¤ä¸ªæ–¹æ³•:
- `__init__(self, kernel_size, num_hiddens)`æ–¹æ³•, æ ¹æ®ç»™å®šçš„å·ç§¯æ ¸å¤§å°å’Œéšè—å±‚å¤§å°, ä»¥ç±»å±æ€§çš„å½¢å¼å®šä¹‰å¿…è¦çš„ç»„ä»¶å’Œå±‚;
- `forward(self, X)`æ–¹æ³•, ç»™å‡ºå‰å‘ä¼ æ’­çš„è¾“å‡º.

### æç¤º
- PyTorchå»ºè®®ä½¿ç”¨[`torch.sigmoid`](https://pytorch.org/docs/stable/generated/torch.sigmoid.html#torch-sigmoid)æ¥å£, è€Œé[`F.sigmoid`](https://pytorch.org/docs/stable/generated/torch.nn.functional.sigmoid.html#torch-nn-functional-sigmoid).
- å·ç§¯æ¨¡å—`ConvModule`çš„è¾“å‡ºå½¢çŠ¶ä¸º`(B, C, H, W)`, åˆ†åˆ«ä»£è¡¨: æ‰¹æ¬¡å¤§å°ã€é€šé“æ•°(ä¸Šè¿°æ¨¡å‹ä¸­ä¸º16)ã€é«˜åº¦(ä¸Šè¿°æ•°æ®ä¸º128)ã€å®½åº¦(ä¸Šè¿°æ•°æ®ä¸º128). ä½ éœ€è¦å°†åä¸‰ä¸ªç»´åº¦â€œå‹æ‰â€ä¸ºä¸€ç»´å†é€å…¥å…¨è¿æ¥å±‚. å¯å‚è€ƒ[`tensor.reshape`](https://pytorch.org/docs/stable/generated/torch.Tensor.reshape.html#torch-tensor-reshape)æ–¹æ³•.

```python
class PizzaOrNotModel(nn.Module):
    def __init__(self, kernel_size=3, num_hiddens=1024):
        super().__init__()
        ### BEGIN YOUR SOLUTION ###
        self.conv_module = ConvModule(kernel_size = kernel_size)
        self.L1 = nn.Linear(16384, 1024)
        self.L2 = nn.Linear(1024, 1)
        self.activation = torch.sigmoid
        
        ### END YOUR SOLUTION ###
    def forward(self, X):
        X = self.conv_module(X).reshape((X.shape[0], -1))
        X = self.L2(F.relu(self.L1(X))).squeeze(1)

        return self.activation(X)
```

```python
pizza_or_not_model = PizzaOrNotModel().to(device)
pizza_or_not_model(X_demo.unsqueeze(0))
```

- å‡½æ•°`train_model_epoch()`ç”¨äºå•ä¸ªå‘¨æœŸçš„è®­ç»ƒ, è¾“å…¥æ¨¡å‹ã€è®­ç»ƒç”¨`DataLoader`å¯¹è±¡ã€ä¼˜åŒ–å™¨`optimizer`ã€æŸå¤±å‡½æ•°`criterion`å’Œè®°å½•æ ‡è®°`log`, è¿›è¡Œå•å‘¨æœŸçš„æ¨¡å‹å‚æ•°ä¼˜åŒ–, è¿”å›è¯¥å‘¨æœŸåœ¨è®­ç»ƒé›†ä¸Šçš„å¹³å‡æŸå¤±å‡½æ•°.
- å‡½æ•°`evaluate_model()`ç”¨äºæ¨¡å‹è¯„ä¼°, è¾“å…¥æ¨¡å‹ã€è¯„ä¼°ç”¨`DataLoader`å¯¹è±¡ã€æŸå¤±å‡½æ•°`criterion`å’Œåˆ†ç±»å†³ç­–é˜ˆå€¼`threshold`, è¿”å›äº”å…ƒå…ƒç»„: å¹³å‡æŸå¤±å‡½æ•°ã€çœŸé˜´æ€§æ ·æœ¬æ•°`tn`ã€å‡é˜³æ€§æ ·æœ¬æ•°`fp`ã€å‡é˜´æ€§æ ·æœ¬æ•°`fn`ã€çœŸé˜³æ€§æ ·æœ¬æ•°`tp`.
- æ³¨æ„: æ¨¡å‹éœ€è¦åœ¨å‡½æ•°å†…å®Œæˆåˆ›å»º, å¹¶ä»¥`.to()`æ–¹æ³•æ¬è¿åˆ°æ­£ç¡®çš„è®¡ç®—è®¾å¤‡ä¸Š.
- [NesterovåŠ¨é‡](https://proceedings.mlr.press/v28/sutskever13.html)æ˜¯ä¸€ç§é€šè¿‡å¼•å…¥å†å²æ¢¯åº¦ä¿¡æ¯åŠ é€ŸSGDä¼˜åŒ–ç®—æ³•æ”¶æ•›çš„æŠ€å·§. åœ¨ç®—æ³•å®è·µä¸­ä¸€èˆ¬æ¨è`momentum`å–åˆå€¼ä¸º0.5.

```python
def train_model_epoch(model, dataloader, optimizer, criterion, log=True):
    model.train()
    train_loss = 0
    num_batches = len(dataloader)
    for i, (X, y_true) in enumerate(dataloader):
        optimizer.zero_grad()
        y_prob = model(X).cuda() if torch.cuda.is_available() else model(X)
        y_true = y_true.cuda() if torch.cuda.is_available() else y_true
        loss = criterion(y_prob, y_true)
        loss.backward()
        optimizer.step()
        train_loss += loss
        if log:
            if i % 100 == 0:
                print(f"Batch No. {i}, loss = {loss.item()}")
    return (train_loss / num_batches).item()
```

```python
@torch.no_grad()
def evaluate_model(model, dataloader, criterion, threshold: float=0.5):
    model.eval()
    loss, tp, fp, tn, fn = 0, 0, 0, 0, 0
    num_batches = len(dataloader)
    for X, y_true in dataloader:
        y_prob = model(X)
        y_prob = model(X).cuda() if torch.cuda.is_available() else model(X)
        y_true = y_true.cuda() if torch.cuda.is_available() else y_true
        loss += criterion(y_prob, y_true)
        y_pred = (y_prob > threshold)
        for idx in range(y_true.shape[0]):
            true, pred = int(y_true[idx].item()), int(y_pred[idx].item())
            if (true, pred) == (1, 1):
                tp += 1
            elif (true, pred) == (1, 0):
                fn += 1
            elif (true, pred) == (0, 1):
                fp += 1
            else:
                tn += 1
    return (loss / num_batches).item(), tn, fp, fn, tp
```

```python
from typing import Dict, List, Tuple
from torch.utils.data import DataLoader
from tqdm import tqdm

def train_model(
    train_subset, val_subset, config: Dict, num_epochs: int=5,
    device="cpu", log=True, save=False
) -> Tuple[List[float], List[float], List[float], List[float]]:
    ### BEGIN YOUR SOLUTION ###
    model = PizzaOrNotModel(kernel_size=config["kernel_size"], num_hiddens=config["num_hiddens"])
    
    if((device != "cpu")&(device != "cuda")):
        print("Invalid device name: 'cpu' or 'cuda'. System exit.")
        sys.exit(-1)
    model.to(device)
    
    train_loader = DataLoader(train_subset, batch_size=config["batch_size"])
    val_loader = DataLoader(val_subset)
    
    train_losses, train_accs, val_losses, val_accs = [], [], [], []
    for i in range(num_epochs):
        if log:
            print(f"Epoch {i+1} starts:")
        # train
        train_loss = train_model_epoch(model=model,
                                       dataloader=train_loader,
                                       optimizer=torch.optim.SGD(model.parameters(), lr=config["lr"], momentum=0.5),
                                       criterion=nn.BCELoss(),
                                       log=log
                                      )
        train_losses.append(train_loss)
        train_acc = 0
        num_samples_train = len(train_subset)
        with torch.no_grad():
            for X, y_true in train_loader:
                y_prob = model(X)
                y_true = y_true.cuda() if torch.cuda.is_available() else y_true
                y_pred = torch.argmax(y_prob, dim=-1)
                y_pred = y_pred.cuda() if torch.cuda.is_available() else y_pred
                train_acc += (y_true == y_pred).sum().item()
            train_accs.append(train_acc / num_samples_train)
    
        if log:
            print(f"Average loss for epoch {i+1}: {train_loss}")
            print("")
        
        # evaluate
        val_loss, tn, fp, fn, tp = evaluate_model(model=model, 
                                                  dataloader=val_loader,
                                                  criterion=nn.BCELoss()
                                                  )
        val_losses.append(val_loss)
        val_acc = (tp + tn) / (tp + tn + fp + fn)
        val_accs.append(val_acc)
        
        
    if save:
        torch.save(model.state_dict(), './PizzaOrNotModel.pt')
        
    return train_losses, train_accs, val_losses, val_accs
    ### END YOUR SOLUTION ###
```
